```python
import pandas as pd
import statsmodels.api as sm
import numpy as np

# Set random seed for reproducibility
np.random.seed(42)

# Number of rows
num_rows = 200

# Generate synthetic data
data = {
    'Age': np.random.randint(20, 65, num_rows),
    'No. of Family Members': np.random.randint(3, 8, num_rows),
    'Total Dependents': np.random.randint(1, 5, num_rows),
    'No. of Earners': np.random.randint(1, 4, num_rows),
    'Income': np.random.randint(30000, 80000, num_rows),
    'Reason for Loan': np.random.randint(1, 5, num_rows),
}

# Create DataFrame
df = pd.DataFrame(data)

# Map Reason for Loan to the specified categories
df['Reason for Loan'] = df['Reason for Loan'].map({
    1: 'Life Expenses',
    2: 'Education',
    3: 'Business',
    4: 'Other'
})

# Add Loan Amount based on a simple linear regression-like relationship
df['Loan Amount'] = 300 + 20 * df['Income'] + 10 * df['Age'] - 5 * df['Total Dependents']


# Define the independent variable (X) and dependent variable (y)
X = df[['Income', 'Age', 'Total Dependents', 'No. of Family Members']]
y = df['Loan Amount']

# Add a constant term to the independent variables
X = sm.add_constant(X)

# Fit the multiple linear regression model
model = sm.OLS(y, X).fit()

# Print the summary statistics
print(model.summary())
```
